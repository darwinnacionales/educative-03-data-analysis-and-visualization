```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def load_flights():
    column_names = ['year', 'month', 'passengers']
    data = pd.read_csv('flights.csv', header=0, names=column_names)
    return data

flights_long=load_flights()

print(flights_long.head())
```

```python
flights = flights_long[flights_long.year >= 1956]

print(flights.head())
```

```python
g = sns.FacetGrid(flights, row="year", margin_titles=True)
g.map(plt.plot, "passengers", color="steelblue")
```

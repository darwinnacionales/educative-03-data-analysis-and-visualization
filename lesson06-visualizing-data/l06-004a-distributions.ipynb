```python
import pandas as pd

def load_boston():
    column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
    data = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
    return data

boston_df = load_boston()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='ticks', palette='Set2')

sns.displot(boston_df.NOX, kde=False)

sns.despine()
```

Here is our same histogram with 100 bins:

```python
sns.set(style='ticks', palette='Set2')

sns.displot(boston_df.NOX, bins=100, kde=False)

sns.despine()
```

You can see the increased variance by increasing the number of bins.

You will notice we have been setting `kde`=False. With this equal to True, it plots the shape of the distribution. We wonâ€™t go into too much detail about this, but you can learn more [here](https://seaborn.pydata.org/tutorial/distributions.html).

```python
sns.set(style='ticks', palette='Set2')

sns.displot(boston_df.NOX, kde=True)

sns.despine()
```

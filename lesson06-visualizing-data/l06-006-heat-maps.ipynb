```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

def load_flights():
    column_names = ['year', 'month', 'passengers']
    data = pd.read_csv('flights.csv', header=0, names=column_names)
    return data

flights_long=load_flights() # Load data from local file

print(flights_long.head())
```

```python
flights = flights_long.pivot(index="month", columns="year", values="passengers")

print(flights.head())
```

```python
f, ax = plt.subplots(figsize=(12, 6))

sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax, cmap='Blues')

# 'coolwarm', 'viridis', 'magma', 'YlGnBu', 'rocket', 'icefire', 'Blues'
```
